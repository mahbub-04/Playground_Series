# 01 - Exploratory Data Analysis (EDA)
## Playground Series S6E1 - Predicting Student Test Scores

**Objectives**
- Understand the dataset structure and features.
- Analyze target variable distribution.
- Explore features relationships and correlations.
- Identify insights for modeling.

```python
playground_series_s6e1_path="/kaggle/input/playground-series-s6e1"
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
train = pd.read_csv(f'{playground_series_s6e1_path}/train.csv')
test = pd.read_csv(f'{playground_series_s6e1_path}/test.csv')
```

```python
print(f'Shape of train data: {train.shape}')
print(f'Shape of test data: {test.shape}')
```

```python
# Display first few rows of training data
train.head()
```

```python
train.columns
```

```python
# Check data types and missing values
train.info()
```

```python
# Basic statistics of numerical features
train.describe()
```

```python
# Check categorical features
categorical_cols = train.select_dtypes(include=['object']).columns
print(f'Categorical columns: {categorical_cols}')
```

```python
for col in categorical_cols :
  print(f'{col} : {train[col].nunique()}')
  print(f'Values: {train[col].unique()}')
```

```python
# Target variable analysis
target = 'exam_score'

print(f'Target variable : {target}')
print(f'Mean :{train[target].mean():.2f}')
print(f'Median :{train[target].median():.2f}')
print(f'Std Deviation :{train[target].std():.2f}')
print(f'Min :{train[target].min():.2f}')
print(f'Max :{train[target].max():.2f}')
```

```python
# Target variable distribution
plt.figure(figsize=(10, 4))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(train[target], bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.title('Distribution of Exam Scores')
plt.axvline(train[target].mean(), color='red', linestyle='--', label=f'Mean: {train[target].mean():.2f}')
plt.legend()

# Boxplot
plt.subplot(1, 2, 2)
plt.boxplot(train[target])
plt.ylabel('Exam Score')
plt.title('Boxplot of Exam Scores')

plt.tight_layout()
plt.show()
```

```python
# Correlation with target
numerical_features = ['age', 'study_hours', 'class_attendance', 'sleep_hours']

correlation_with_target = train[numerical_features + [target]].corr()[target].sort_values(ascending=False)
print("Correlation with Exam Score:")
print(correlation_with_target)
```

## Key Findings from EDA:

### Dataset Overview:
- Training samples: 630,000
- Features: 12 (6 numerical + 6 categorical)
- Target: exam_score (continuous, range 19.6 to 100)
- No missing values

### Target Distribution:
- Mean: 62.51, Median: 62.60 (nearly equal -> symmetric)
- Bell shaped / Normal distribution
- Good for regression modeling

### Feature Correlation with Target:
1. **study_hours: 0.76** (strongest predictor!)
2. class_attendance: 0.36 (moderate)
3. sleep_hours: 0.17 (weak)
4. age: 0.01 (negligible)
